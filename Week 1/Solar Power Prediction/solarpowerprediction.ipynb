```python
import matplotlib.pyplot as plt
```

```python
# Plot distribution of power
plt.figure(figsize=(10,6))
sns.histplot(df['generated_power_kw'], bins=30, kde=True)
plt.title('Distribution Of Generated Power (kW)')
plt.xlabel('Generated Power (kW)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['wind_direction_10_m_above_gnd'], bins=30, kde=True, color='red')
plt.title('Wind Direction 10m above ground')
plt.xlabel('Wind Direction (0° to 360°)')
plt.ylabel('Time (Hours)')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['temperature_2_m_above_gnd'], bins=30, kde=True, color='black')
plt.title('Temperature 10m above ground')
plt.xlabel('Degree Celsius (°C)')
plt.ylabel('Time (Months)')
plt.show()
```
